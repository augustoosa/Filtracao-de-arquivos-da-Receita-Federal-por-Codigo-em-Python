### FILTRANDO O ARQUIVO E SALVANDO O RESULTADO FINAL

```python
# Caminho do segundo arquivo CSV
file_path_empresas = 'Empresas9/Emp9.csv'


# Tenta ler o segundo arquivo CSV com opções adicionais para lidar com erros
df_empresas = pd.read_csv(file_path_empresas,encoding='latin1', delimiter=';')

lista=["CNPJ BÁSICO", "RAZÃO SOCIAL / NOME EMPRESARIAL", "NATUREZA JURÍDICA", "QUALIFICAÇÃO DO RESPONSÁVEL", "CAPITAL SOCIAL DA EMPRESA", "PORTE DA EMPRESA", "ENTE FEDERATIVO RESPONSÁVEL"]
df_empresas.columns=lista
# Filtra as linhas do segundo arquivo que correspondem aos CNPJs da lista
df_resultado = df_empresas[df_empresas['CNPJ BÁSICO'].isin(cnpj_basico)]

# Salva as informações filtradas em um novo arquivo CSV
output_path = 'resultado9.csv'
df_resultado.to_csv(output_path, index=False)

df_resultado
```

# JUNTANDO TODOS OS RESULTADOS EM 1 SÓ

```python
import pandas as pd
import glob

# Lista para armazenar todos os DataFrames
dataframes = []

# Loop para ler todos os arquivos csv
for i in range(10):
    file_name = f'resultado{i}.csv'
    df = pd.read_csv(file_name)
    dataframes.append(df)

# Concatenar todos os DataFrames
combined_df = pd.concat(dataframes, ignore_index=True)

# Salvar o DataFrame combinado em um novo arquivo CSV
combined_df.to_csv('resultado_combinado.csv', index=False)
```
